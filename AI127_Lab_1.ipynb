```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("Salary_Data.csv")
```

```python
ds.head()
ds.shape
```

```python
x = ds.iloc[:, :1]
y = ds.iloc[:, 1]
```

```python
xtrain = x[:20]
ytrain = y[:20]

xtest = x[21:]
ytest = y[21:]

xtest, ytest = xtest.to_numpy(), ytest.to_numpy()
```

```python
def regressor(xtrain, ytrain):
    ytrain = ytrain.to_numpy()
    xtrain = xtrain.to_numpy()
    yavg = np.mean(ytrain)
    xavg = np.mean(xtrain)
    
    p = 0
    q = 0
    
    for i in range(20):
        p = p + (xtrain[i]*ytrain[i])
        q = q + (xtrain[i]**2)
    Sxy = (p/20) - (xavg*yavg)
    Sxx = (q/20) - (xavg**2)
    
    B1 = Sxy/Sxx
    B0 = yavg - (B1*xavg)
    return B0, B1
```

```python
b0, b1 = regressor(xtrain, ytrain)
```

```python
yline = (x.to_numpy()*b1) + b0

plt.plot(x, yline)
plt.scatter(x, y.to_numpy(), color = 'red')
plt.scatter(xtest, ytest, color = 'blue')
plt.legend(['Linear regressor', 'Train data', 'Test data'])
plt.show()
```
